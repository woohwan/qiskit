### Task 1.1 Pauli Operators
#### Objective: Define Pauli Operators
**overview**: This notebook covers Pauli operators, how to initialize them, their attributes and methods.

Pauli operators are fundamental building blocks in quantum computing that represent single-qubit and multi-qubit operations.  
In Qiskit, the `Pauli` class provide various ways to create and manuplate these operations

```python
# Setup: Import nessesary libraries
from qiskit.quantum_info import Pauli, ScalarOp
import numpy as np
from qiskit import QuantumCircuit

print("Libraries imported successfully.")
```

#### Initialization Methods
##### 1. Pauli String Representation
Pauli operators can be created from string representaions using Pauli matrices

- I: Identity operator
- X: Pauli-X (bitflip) operator
- Y: Pauli-Y operator
- Z: Pauli-Z (phase-flip) operator

Note: Qiskit uses Little Endian Notation where qubit(n-1) is the leftmost and  
qubit 0 is the rightmost in the string representation.
The phase can be specified as `  `(default), `-`, i, or `-`

```python
# Create Pauli operators from string representation
p = Pauli("XZ")     # No phase (default +1)
p2 = Pauli('iYY')   # with imaginary phase i
p3 = Pauli('-iZX')  # with negative imaginary phase

print(f"P: {p}, P2: {p2}, P3: {p3}")
```

#### 2. Boolean Binary Representation
Pauli operators can also be created using boolean numpy arrays:
- `z` array: Specifies Z components (True = Z operator, False = I operator)
- `x` array: Specifies X components (True = X operator, False = I operator)

The actual Pauli operator at each qubit position is determined by:
- (z=False, x=False) -> I (identity)
- (z=False, x=True) -> X
- (z=True, x=False) -> Z
- (z=True, x=True) -> Y (since Y= iXZ)

```python
# Define Z and X component using boolean array
# I on qubit 0, X on qubit 1, Z on qubit 2, Y on qubit 3
z = np.array([False, False, False, True])
x = np.array([False, True, False, True])

pauli_array_rep = Pauli((z, x))
print(f"Pauli Operator from Boolean Array: {pauli_array_rep}")
```
